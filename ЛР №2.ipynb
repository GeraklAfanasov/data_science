```python
import pandas as pd

# Загрузим данные из 'NewRecorder.csv'
df = pd.read_csv('NewRecorder.csv', sep=';')

# Отфильтруем числовые столбцы
numeric_data = df.select_dtypes(include='number')

# Выполним описанные статистические операции
statistics = pd.DataFrame(index=['count', 'sum', 'mean', 'median', 'min', 'max', 'mode', 'abs', 'std'])

for col in numeric_data.columns:
    stats = []
    stats.append(numeric_data[col].count())
    stats.append(numeric_data[col].sum())
    stats.append(numeric_data[col].mean())
    # stats.append(numeric_data[col].mad())
    stats.append(numeric_data[col].median())
    stats.append(numeric_data[col].min())
    stats.append(numeric_data[col].max())
    stats.append(numeric_data[col].mode().values[0])
    stats.append(numeric_data[col].abs().mean())
    stats.append(numeric_data[col].std())

    statistics[col] = stats

# Выведем результат
print(statistics)
```
